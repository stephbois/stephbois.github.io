```python
class iDataAnalysis:
    '''Algorithms for Data Analysis on input dataframe (df)'''
    def __init__(self, dataframe):
        self._dataframe = dataframe

#1. Implement an algorithm to identify the top 10 subcategories by sales

    #Algorithm top_ten was made generalizable for the user to be able to group the top 10 of the defined variable x by a choden group
    #To answer question 1, the user could call the method using the dataset's sales column as variable x, and the dataset's subcategory column as the group
    def top_ten (self, x, group):
        '''Summation grouping of "x" by "group".
        Returns a dictionary of the top ten results of x for each group'''
        #Initializing list of dictionary key values pairs of key("group") : value("x")
        group = tuple(group)
        x = list(x)
        my_list = []
        for i in range(len(group)):
            my_dict = {"group": group[i], "x": x[i]}
            my_list.append(my_dict)
        #Group and sum values("x") by key("group")
        group_keys = "group"
        sum_keys = "x"
        result = {}
        for i in my_list:
            subgroup = i[group_keys]
            if subgroup not in result:
                for x in sum_keys:
                    result[subgroup] = int(0)
            for n in sum_keys:
                result[subgroup] += int(i[n])
        #Sort results in descending order and return the top 10 results
        #Python's sorted() function uses Timsort which is a hybrid of merge sort and insertion sort
        name = lambda x: (x[1], x[0])
        sorted_result = sorted(result.items(), key = name, reverse = True)
        return (sorted_result[0:10])

    #Algorithm top_sales_by_subcategory is another sample algorithm but made specific to grouping sales by subcategory
    def top_sales_by_subcategory (self):
        '''Summation grouping of sales by subcategory.
        Returns the top ten sales by group'''
        #Group by sub category an sum number of sales
        sumSales = df.groupby(['sub_category'])['sales'].sum().sort_values(ascending = False)
        return(sumSales[0:10])

#2. Design a data structure to store the monthly sales trend for the entire superstore.
    def store_monthly_sales(self):
        '''Stores the monthly sales trend (both "sales" and "profit") into a list of lists'''
        #Store month/year values into list
        month_year = df.month_year
        month_year_list = []
        for i in range(len(month_year)):
            if month_year[i] not in month_year_list:
                month_year_list.append(month_year[i])
        #Group and sum monthly sales and profit and store values into lists
        monthlySales = df.groupby(['year','month'])['sales', 'profit'].sum()
        sales_list = monthlySales['sales'].tolist()
        profit_list = monthlySales['profit'].tolist()
        #Store values in a list of lists
        col_list = []
        row_list = []
        for i in range(len(month_year_list)):
            col_list = month_year_list[i], sales_list[i], profit_list[i]
            row_list.append(col_list)
        return row_list

#3. Implement a search algorithm to find all days where a promo led to sales above a certain threshold.
    def searchDays(self, treshold):
        '''Search for days where a promotional item led for sales above selected sales "treshold".
        Returns tuple(days, item, sales, and discount) above selected sales treshold'''
        treshold = int(treshold)
        #Linear Search for position of treshold and store sales above treshold
        above_treshold_position = []
        position = 0
        while position < len(df.sales):
            if treshold < df.sales[position]:
                if df.discount[position] > 0:
                    above_treshold_position.append(position)
            position += 1
        my_lyst = []
        for i in above_treshold_position:
            my_tuple = (df.order_date[i], df.product_name[i], df.sales[i], df.discount[i])
            my_lyst.append(my_tuple)
        return my_lyst
```

**Optimizing for memory efficiency:**

Memory efficiency of the code can be optimized by storing data into data-structures and collections
that are memory efficient (such as those that are) and by reducing the size of objects in the memory. Arrays are data structures which are given a determined capacity upon declaration and are allocated a block of adjacent memory cells for that capacity. Arrays can threfore be sized and resized accordingly to avoid wasting memory by checking the load factor. An array with a low load factor for example (0.25) means that the logical size of the array is much smaller than its physical size, meaning that the array is taking up more memory space but only storing a small ammount of data. To be memory efficient, we would want to create a function to resize and decrease
the physical size of array so that it doesn't take up as much memory. This is another example of a tradeoff of memory and time efficiency, because the transfer of elements when resizing the arrays takes up time.

Question 2 asked that I create a data structure for storing to the monthly sales trend for the entire superstore. The original structure was a list of lists. To store the elements in a data structure that is more memory efficient, we will store the monthly sales in an array of lists. To document the changes in memory efficiency, we will check the memory size taken up by the data structure using *getsizeof* from *sys*.   

```python
#2. Designing an array class to store the monthly sales trend
class iArray():
    '''A representation of an array'''
    def __init__(self, capacity, fillValue = None):
        '''Capacity is the size of the array.
        None object will be placed at each position.'''
        self._items = list()
        for count in range(capacity):
            self._items.append(fillValue)

    def __len__(self):
        #Obtaining the size of the array
        return len(self._items)

    def __str__(self):
        #Representing the array as String
        return str(self._items)

    def __setitem__(self, index, newItem):
        #Replacing an element or assigning a value to an element
        self._items[index] = newItem

    def __getitem__(self, index):
        #Obtaining an element in the array
        return self._items[index]

    def __iter__(self):
        #Traversing elements with loops
        return iter(self._items)

    def loadFactor(self):
        '''Returns the load factor of the array'''
        logicalSize = len(([x for x in self if x is not None]))
        physicalSize = len(self)
        loadFactor = logicalSize/physicalSize
        return(loadFactor)
```

**Optimizing for time efficiency:**

Code for the **searchDays** method of the **iDataAnalysis** class uses a linear search model of O(n). The search iterates through every single item of the list to determine which item positions of 'sales' are above the treshold. The O(n) search model is less time efficient for lists with a large *n* value since the model must iterate through *n* number of items during the search. Using a binary search algorithm on a sorted list, such as using Python's bisect library, can help us identify the position of items above the treshold with a worst case scenario logarithmic O(log2(n)) search. This logarithmic search greatly reduces the amount of steps needed to conduct the search on lists or large value n.

To help document changes in time efficiency, we will time both the older version of the algorithm and the newer version using default_timer from timeit.

```python
#Optimizing question number 3 for time efficiency (using Python's bisect library binary search algorithm)
import bisect

#3. Implement a search algorithm to find all days where a promo led to sales above a certain threshold.
def searchDaysOptimized (treshold):
        '''Search for days where a promotional item led to sales above "treshold".
        Returns tuple(days, item, sales, and discount)'''

        #Sort dataframe by sales
        sorted_sales = df.sort_values(['sales'], ascending= True)

        #Remove rows withouth a discount
        sorted_sales = sorted_sales[sorted_sales['discount'] > 0]

        #Store values with discounts in list to assign new index position and iterate through list:
        sales_list = list(sorted_sales.sales)
        date_list = list(sorted_sales.order_date)
        product_list = list(sorted_sales.product_name)
        promotion_list = list(sorted_sales.discount)

        #Use python bisect for binary search for position of treshold
        treshold_position = bisect.bisect_right(sales_list, treshold)

        #Store data, sales, product, and discount into new list of lists
        my_lyst = []
        i = treshold_position
        while i < len(sales_list):
            my_tuple = (date_list[i], product_list[i], sales_list[i], promotion_list[i])
            my_lyst.append(my_tuple)
            i += 1
        return my_lyst
```
